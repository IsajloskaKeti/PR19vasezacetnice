```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('master.csv')
```

```python
df = pd.DataFrame(data1, columns = ['country']).country.unique()
print(df)
len(df)  #This dataset includes data for 101 country
```

```python
#In these 10 countries are made the most suicides.
data1.groupby(['country']).suicides_no.sum().nlargest(10).plot(kind='bar',figsize=(30,30))  
```

```python
#Relationship between sex with made suicides by year.
data1.groupby(['sex','year']).suicides_no.sum().plot(kind='bar',figsize=(30,30))
```

```python
count1 = sns.barplot(x='generation', y='suicides_no', hue='sex', data=data1,)
print(count1) #Relation between generation,sex and made suicides
```

```python
#Number of suicides globally by year
suicidesperyear = pd.DataFrame(data1['suicides_no'].groupby(data1['year']).sum())
suicidesperyear = suicidesperyear.reset_index().sort_index(by='suicides_no',ascending=False)
suicidesperyear2 = suicidesperyear
fig = plt.figure(figsize=(40,40))
plt.title(' Number of suicides globally')
sns.set(font_scale=5)
sns.barplot(y='suicides_no',x='year',data=suicidesperyear2,palette="Reds");
plt.ylabel('Number of made suicides')
plt.xlabel('')
plt.xticks(rotation=60)
plt.tight_layout()
```

```python
dataslovenia=data1.query("country=='Slovenia'")
suicidesperyear = pd.DataFrame(dataslovenia['suicides_no'].groupby(dataslovenia['year']).sum())
suicidesperyear = suicidesperyear.reset_index().sort_index(by='suicides_no',ascending=False)
suicidesperyear2 = suicidesperyear
fig = plt.figure(figsize=(50,40))
plt.title('Count of suicides for Slovenia from 1995 to 2016')
sns.set(font_scale=6)
sns.barplot(y='suicides_no',x='year',data=suicidesperyear2,palette="GnBu_d");
plt.ylabel('Count of made suicides')
plt.xlabel('')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
#Number of made suicides globally by sex
suicidesperyear = pd.DataFrame(data1['suicides_no'].groupby(data1['sex']).sum())
suicidesperyear = suicidesperyear.reset_index().sort_index(by='suicides_no',ascending=False)
suicidesperyear2 = suicidesperyear
fig = plt.figure(figsize=(40,30))
plt.title(' Number of suicides globally by sex')
sns.set(font_scale=4)
sns.barplot(y='suicides_no',x='sex',data=suicidesperyear2,palette="Reds");
plt.ylabel('Number of made suicides')
plt.xlabel('')
plt.xticks(rotation=60)
plt.tight_layout()
```
